```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from skimage import io

# Load the image
image = io.imread(r"C:\Users\Bindushree\OneDrive\Pictures\Camera imports\2024-10-28\IMG_20231217_161823.jpg")

# Apply Gaussian filter
gaussian_blur = gaussian_filter(image, sigma=2)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gaussian Blurred Image')
plt.imshow(gaussian_blur)
plt.axis('off')

plt.show()
```
